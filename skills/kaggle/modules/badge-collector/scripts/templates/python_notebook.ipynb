```python
import pandas as pd
import numpy as np

# Generate sample data
np.random.seed(42)
n = 100
data = pd.DataFrame({
    'x': np.random.randn(n),
    'y': np.random.randn(n),
    'label': np.random.choice(['A', 'B', 'C'], n)
})

print(f'Data shape: {data.shape}')
print(data.describe())
```

```python
# Group by label
grouped = data.groupby('label').agg(['mean', 'std'])
print(grouped)
```
